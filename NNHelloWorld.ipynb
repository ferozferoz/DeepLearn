```
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import numpy as np
```

```
#There is one layer and one connected unit represented by Dense
model = Sequential([Dense(units=1, input_shape=[1])])
# error function determines the error while the SGD stochastic gradient descent determines who good was the estimate and how much the parameters of network to be modified. 
model.compile(optimizer='sgd', loss='mean_squared_error')
```

```
# we have created numpy array of the input and output data
x= np.array([-5,-2,0,1,4,7], dtype=float)
y=np.array([-9,-3,1,3,9,15], dtype=float)
```

```
import matplotlib.pyplot as plt 
plt.plot(x,y)
# plot indicates its a linear relationship
```

```
history = model.fit(x,y,epochs=500)
```

```
# we were suppose to print 17 but could only print 16.99. Why ? 
print(model.predict([8.0]))

#Ans - It is because we have few examples with answers which was used by the machine to learn parameters. total of six datapoint were used by NN. Although there is a linear relationship between numbers but 
# NN training does not start with that belief. It start with a random value and keep converging toward 2x+1 but its not exactly 2x+1. As we can see below the loss is almost 0 but not equal to 0.    
```

```
plt.plot(history.history['loss'])
```
